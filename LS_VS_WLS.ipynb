```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf

plt.rcParams["figure.figsize"] = (12, 9)
```

```python
fin_data_set = pd.read_csv("/content/fin_data.csv")
```

```python
forex_data_set = pd.read_csv("/content/forex.csv")
```

```python
#OLS Model finance data
model_1 = smf.ols(
    "Coke_Q_EX_R ~ Dow_Q_EX_R",
    data=fin_data_set,
).fit()
model_1.summary()
```

```python
# WLS model

# Add Absolute residuals and fitted values to dataset columns
fin_data_set["abs_residuals"] = np.abs(model_1.resid)
fin_data_set["fitted_values"] = model_1.fittedvalues

# Fit OLS model with absolute residuals and fitted values
model_temp = smf.ols("abs_residuals ~ fitted_values", data=fin_data_set).fit()

# Compute weights and add it to the data_set column
weights = model_temp.fittedvalues
weights = weights**-2
fin_data_set["weights"] = weights

# WLS model
Y = fin_data_set["Coke_Q_EX_R"].tolist()
X = fin_data_set["Dow_Q_EX_R"].tolist()
X = sm.add_constant(X)  # add a intercept point

model_WLS = sm.WLS(Y, X, fin_data_set["weights"]).fit()
model_WLS.summary()
```

```python
#OLS Model forex data
model_2 = smf.ols(
    "DXY ~ X10Y_TBY",
    data=forex_data_set,
    ).fit()
model_2.summary()
```

```python
# WLS model

# Add Absolute residuals and fitted values to dataset columns
forex_data_set["abs_residuals"] = np.abs(model_2.resid)
forex_data_set["fitted_values"] = model_2.fittedvalues

# Fit OLS model with absolute residuals and fitted values
model_temp = smf.ols("abs_residuals ~ fitted_values", data=forex_data_set).fit()

# Compute weights and add it to the data_set column
weights = model_temp.fittedvalues
weights = weights**-2
forex_data_set["weights"] = weights

# WLS model
Y = forex_data_set["DXY"].tolist()
X = forex_data_set["X10Y_TBY"].tolist()
X = sm.add_constant(X)  # add a intercept point

model_WLS = sm.WLS(Y, X, forex_data_set["weights"]).fit()
model_WLS.summary()
```
