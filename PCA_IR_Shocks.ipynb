```python
from openbb import obb
import pandas as pd
import numpy as np
```

```python
rates = (obb.fixedincome.government.treasury_rates(start_date="2020-01-01")
          .to_df()
          .dropna()
          .div(100)
        )
rates
```

```python
C = rates.cov()
```

```python
eigenvalues, eigenvectors = np.linalg.eig(C)
lambda_sqrt = np.sqrt(eigenvalues)
eigv_decom = np.diag(lambda_sqrt)
```

```python
B = eigv_decom @ eigenvectors.T
B = pd.DataFrame(
    data=B[:4] * 100,
    index=["wiggle", "flex", "twist", "shift"],
    columns=rates.columns
)
```

```python
B
```

```python
random_shocks = np.random.normal(0, 1, size=(4, ))
random_shocks @ B
```
